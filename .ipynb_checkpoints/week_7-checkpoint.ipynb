```python
import bs4
import requests
import re
import json

url = "https://www.elgiganten.dk/catalog/gaming/dk-gaming-laptop/gaming-laptop"

r = requests.get(url)

soup = bs4.BeautifulSoup(r.text, 'html.parser')


cells = soup.select('.col-infos')
titles = []
prices=[]

for e in cells:
    name = e.select(".product-name")
    price = e.select(".product-price")
    for n in name:
        titles.append(n.text.strip())
    for pr in price:
        txt = pr.text.strip()
        
        pris = re.compile(r'^\d+\s+\d+')
        m = pris.search(txt)
        prices.append(m.group(0).replace(u'\xa0', ''))
        
    
products=dict(zip(titles, prices))
products = {k: v for k, v in sorted(products.items(), key=lambda item: item[1])}
print(products)

print("------------------------------------------------")

string = json.dumps(products, ensure_ascii=False)
print(string)
```

```python
import ast
import operator
import matplotlib.pyplot as plt

asuses = re.compile(r'\"Asus.*?\d{3}\"')

ass = asuses.findall(string)
print(ass)
anus = ", ".join(ass)

print(anus)


anus_dict = ast.literal_eval("{"+anus+"}")
print("non sorted")
print(anus_dict)
for value in anus_dict:
    anus_dict[value] = int(anus_dict[value])
print(anus_dict)
anus_dict2 = dict( sorted(anus_dict.items(), key=operator.itemgetter(1)))

print("sorted list")
print(anus_dict2)
anus_items = anus_dict2.items()
first_item = list(anus_items)[:1]
(tit, pri) = first_item[0] 

print(tit)
print(pri)


plt.xticks(rotation=90)
plt.bar(anus_dict2.keys(),anus_dict2.values())
plt.show()
```
